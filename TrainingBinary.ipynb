```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Activation, Dropout, Flatten, Dense
from keras.layers import LeakyReLU
import tensorflow as tf
from keras.optimizers import SGD
from keras.callbacks import LearningRateScheduler


model = Sequential()

model.add(Conv2D(32, (3, 3), input_shape=(50, 100,3)   ))
#model.add(LeakyReLU(alpha=0.1))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2) ,strides = (2,2) ))
#model.add(Dropout(0.5))

model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2) ,strides = (2,2) ))
#model.add(Dropout(0.5))

model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2) ,strides = (2,2)  ))
#model.add(Dropout(0.5))


#model.add(Conv2D(64, (3, 3), activation='relu'))
#model.add(MaxPooling2D((2, 2) ,strides = (2,2)  ))
#model.add(Dropout(0.5))

model.add(Flatten())
model.add(Dense(64))
#model.add(LeakyReLU(alpha=0.1))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(1))
model.add(Activation('sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])
model.summary()
```

```python
###RUN SLIDING WINDOW

def sliding_window(im , w , h , stride):
    M = im.size[1]
    N = im.size[0]
    ans = []
    i = 0
    while i < (M - h + 1):
        print(i)
        j = 0
        
        while j < (N - w + 1):
            conf  = loaded_model.predict(np.asarray(im.crop((j,i,j+w,i+h)).resize((100,50))).reshape(1,50,100,3))
            if conf > 0.98 :
                ans.append((i,j,i+h,j+w))
                
            j+=stride
                
        i+=stride
    return ans
ans  = sliding_window(im , 200 , 67 , 10)
```

```python
#### OUTPUT
for rr in ans:
    data[rr[0]:rr[2],rr[1]-1:rr[1]+1 , :] = [255.0 , 0.0 , 0.0]
    data[rr[0]:rr[2],rr[3]-1:rr[3]+1 , :] = [255.0 , 0.0 , 0.0]
    
    data[rr[0]-1:rr[0]+1,rr[1]:rr[3] , :] = [255.0 ,0.0 , 0.0]
    data[rr[2]-1:rr[2]+1,rr[1]:rr[3] , :] = [255.0,0.0,0.0]
    
pyplot.imshow(data)
pyplot.show()
```

```python
#### VISUALIZATION###
from keras import models
layer_outputs = [layer.output for layer in loaded_model.layers[:9]]
activation_model = models.Model(inputs=loaded_model.input, outputs=layer_outputs)

im = Image.open("usimages/usimages/car-10.png")
im = im.resize((512,512))
im = im.crop((180,155,280 , 197))
im = im.resize((100,50))
data = np.asarray(im).copy()
data = data.reshape(1,50,100,3)
activations = activation_model.predict(data)
```

```python
from matplotlib import pyplot as plt
layer_names = []
for layer in loaded_model.layers[:9]:
    layer_names.append(layer.name) # Names of the layers, so you can have them as part of your plot
    
images_per_row = 16

for layer_name, layer_activation in zip(layer_names, activations): # Displays the feature maps
    n_features = layer_activation.shape[-1] # Number of features in the feature map
    size = layer_activation.shape[1]#The feature map has shape (1, size, size, n_features).
    size2 = layer_activation.shape[2]
    n_cols = n_features // images_per_row # Tiles the activation channels in this matrix
    display_grid = np.zeros((size * n_cols, images_per_row * size2))
    for col in range(n_cols): # Tiles each filter into a big horizontal grid
        for row in range(images_per_row):
            channel_image = layer_activation[0,
                                             :, :,
                                             col * images_per_row + row]
            channel_image -= channel_image.mean() # Post-processes the feature to make it visually palatable
            channel_image /= channel_image.std()
            channel_image *= 64
            channel_image += 128
            channel_image = np.clip(channel_image, 0, 255).astype('uint8')
            display_grid[col * size : (col + 1) * size, # Displays the grid
                         row * size2 : (row + 1) * size2] = channel_image
    scale = 1. / size
    plt.figure(figsize=(scale * display_grid.shape[1],
                        scale * display_grid.shape[0]))
    plt.title(layer_name)
    plt.grid(False)
    plt.imshow(display_grid, aspect='auto', cmap='viridis')
    plt.show()
```
